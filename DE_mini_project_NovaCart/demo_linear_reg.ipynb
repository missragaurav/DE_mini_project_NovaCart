```python
import pandas as pd
import numpy as np
```

```python
data = {
    "x1" : [4,9,8,8,7,5],
    "x2" : [1,2,3,1,1,0],
    "x3" : [3,1,2,1,4,2],
    "y"  : [52,51,50,52,70,60]
}

df = pd.DataFrame(data)
df
```

```python
import math

# Train Test split

def train_test_split(data, tar):
    x = len(data['x1'])
    idx = math.floor(0.7*x)

    train = pd.DataFrame()
    test = pd.DataFrame()

    for col in data:
        train[col] = data[col][:idx]
        test[col] = data[col][idx:]
    
    y = test[tar]
    test = test.drop(columns=[tar]) 

    return train, test, y
```

```python
tar = 'y'
train_data, test_data, y = train_test_split(df, tar)

print('Training Data : \n', train_data)
print("Testing Data : \n", test_data)
print('Target test values : \n', y)
```

```python
# Regression model

def linear_regression(df, x, y):
    x_mean = df[x].mean()
    y_mean = df[y].mean()
    
    # Slope (m)
    numerator = np.sum((df[x] - x_mean) * (df[y] - y_mean))
    denominator = np.sum((df[x] - x_mean) ** 2)
    slope = numerator / denominator
    
    # Constant (c)
    constant = y_mean - (slope * x_mean)
    
    return slope, constant
```
